```python
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np
import wclablib as wclab
import seaborn as sbn
from scipy import signal
```

```python
#dat = np.loadtxt('/nas-6000/wclab/James/AMOG/20150520/Data/t12_h7.6_w3.9_90_08_02.dat',delimiter=',')
dat = np.loadtxt('/nas-6000/wclab/James/AMOG/20150520/Data/t8_h7.1_w4.8_90_08_02.dat',delimiter=',')
time = dat[:,0]; lift = dat[:,1]
plt.plot(time,lift);
```

```python
f,mx,t = wclab.psdd(100,lift)
plt.plot(f,mx)
plt.xlabel('frequency (hz)')
plt.ylabel('Power')
plt.xlim(0,5)
```

```python
F,MX,T = wclab.psdd3(100,lift,1024,16)

plt.pcolormesh(T,F,MX)
plt.ylim(0,4);
plt.clim(0,1e-3)
plt.xlabel('Time (s)')
plt.ylabel('Frequency (hz)')
```

```python
widths = np.arange(0.1,100,0.1)
cwtmatr = signal.cwt(lift,signal.ricker,widths)
```

```python
plt.imshow(cwtmatr,cmap='PRGn', aspect='auto')
plt.show()
```
