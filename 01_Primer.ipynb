## How to use a notebook?

### Installation:

#### On linux:

1. Download python 2.7.10
2. Download python 3.4.3

    pip install ipython

#### On windows / mac:
1. Download anaconda

        conda update conda
        conda update ipython ipython-notebook ipython-qtconsole



### Running:

#### To run locally:
    ipython notebook
#### To run server:
1. Create a SSL certificate
        openssl req -x509 -nodes -days 365 -newkey rsa:1024 -keyout mycert.pem -out mycert.pem
2. Secure notebook server with a password:
        ipython
        from IPython.lib import passwd
        passwd()
3. Create a server profile
        ipython profile create nbserver
4. Edit that profile as follows, edit ```ipython_notebook_config.py```
        c.NotebookApp.certfile = u'path to mycert.pem'
        c.NotebookApp.ip = '*'
        c.NotebookApp.open_browser = False
        c.NotebookApp.password = u'[your hashed password here]'
        c.NotebookApp.port = 9999
5. Launch notebook with
        ipython notebook --profile=nbserver
6. Must browse to https://IP:PORT
